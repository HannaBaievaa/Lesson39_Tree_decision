```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import graphviz
```

```python
from sklearn import datasets
iris=datasets.load_iris()
iris
```

```python
#задаємо предикторні змінні(X-на них навчатиметось модель та у-цільова змінна,яку визначатиме модель)
X,y = iris.data, iris.target
```

```python
#розбиваємо датасет на тренувальну та тестувальну вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state=42)
```

```python
#створюємо екземпляр моделі
clf_model = DecisionTreeClassifier(criterion="gini", random_state=42, max_depth=3, min_samples_leaf=5)
```

```python
#навчаємо нашу модель
clf_model.fit(X_train,y_train)
```

```python
#зробимо прогноз, тобто виконаємо предикцію 
y_predict = clf_model.predict(X_test)
```

```python
#оцінимо якість зробленого прогнозу за допомогою метрики accuracy
metrics.accuracy_score(y_test,y_predict)
```

### Візуалізація дерева 

```python
df_target = pd.DataFrame(y, columns=list('y'))
target_df = list(df_target['y'].unique())
target=[]
for i in target_df:
    target.append(str(i))

df_data = pd.DataFrame(X, columns=['x1','x2','x3','x4'])
feature_names = list(df_data.columns)
```

```python
dot_data = tree.export_graphviz(clf_model,feature_names=feature_names,class_names=target)  
graph = graphviz.Source(dot_data)
graph
```

```python
#розрахуємо значимість параметрів
importances = clf_model.feature_importances_
importances
```

```python
#підпишемо та відсортуємо параметри
df_train = pd.DataFrame(X_train, columns=['x1','x2','x3','x4'])

sorted_importances = pd.DataFrame({'feature': list(df_train.columns),
                   'importance': clf_model.feature_importances_}).\
                    sort_values('importance', ascending = False)
sorted_importances
```
